```python
from tabulate import tabulate
from pandas import DataFrame


xnc=[]
xn=[]
ui=[]
rn=[]
longitud=[]
semillacuadrada=0

iteraciones=int(input('Cantidad de Iteracciones: '))
semilla=int(input('Valor de la semilla: '))
digitos=int(input('Digitos a considerar: '))

for i in range(iteraciones):
    scuadrada=semilla*semilla
    xn.append(semilla)
    xno=str(scuadrada)
    longit=len(xno)
    longitud.append(longit)
    ui1=int((int((longit/2))-int((digitos/2)))+1)
    ui2=int((int((longit/2))+int((digitos/2)))+2)
    semillanueva=xno[int(ui1):int(ui2)]
    snu=int(semillanueva)
    uno="1"
    for j in range(digitos):
        uno=uno+"0"
    RN=(snu/int(uno))
    semilla=snu
    ui.append(snu)
    xnc.append(scuadrada)
    rn.append(RN)


dt = {'xn' : xn, 'xn*xn' : xnc,'longitud' : longitud,'ui' : ui,'rn' : rn,}
df= DataFrame(dt)
df
```
